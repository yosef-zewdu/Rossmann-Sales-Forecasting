**Short description:**

- Store: a unique ID for each store
- Sales: the turnover for any given day.
- Customers: the number of customers on a given day.
- Open: an indicator for whether the store was open: 0 = closed, 1 = open.
- Promo: indicates whether a store is running a promo on that day.
- StateHoliday: indicates a state holiday. Normally all stores, with few exceptions, are closed on state holidays. 
- SchoolHoliday: indicates if the (Store, Date) was affected by the closure of public schools. a = public holiday, b = Easter holiday, c = Christmas, 0 = None

## Handling missing values in store 

```python
# missing values
store.isnull().sum()
```

```python
# missing values in CompetitionDistance
store[pd.isnull(store.CompetitionDistance)]
```

```python
# fill NaN with a median value (skewed distribuion)
store['CompetitionDistance'].fillna(store['CompetitionDistance'].median(), inplace = True)
```

```python
#  promo2 
promo2null = store[pd.isnull(store.Promo2SinceWeek)]  # the 544 stores that has no promo2sinceweek 
promo2null[promo2null.Promo2 != 0].shape              # check how many stores have promo2 
```

- 544 stores don't have promo2. so there's no information about `Promo2SinceWeek`, `Promo2SinceYear`, `PromoInterval` .We can replace these values by zeros. 

- The same goes for `CompetitionOpenSinceMonth` and `CompetitionOpenSinceYear` since we don't have information about them, we replace them by zeros

```python
# Replace by zeroes 
store.fillna(0, inplace = True)
```

## Feature Engineering  

```python
train['Date'] = pd.to_datetime(train['Date'])
# train.set_index('Date', inplace=True)
```

```python
# data extraction
train['Year'] = train['Date'].dt.year
train['Month'] = train['Date'].dt.month
train['Day'] = train['Date'].dt.day
train['WeekOfYear'] = train['Date'].dt.isocalendar().week 
```

```python
# adding Sales per customer
train['SalePerCustomer'] = train['Sales']/train['Customers']
```

```python
# add is_holiday to check whether the day is holiday or not 

#  Define which StateHoliday values are considered holidays
holiday_values = ['a', 'b', 'c']

# Create a new column to compare
train['is_holiday'] = train['StateHoliday'].isin(holiday_values).astype(int)
```
